```python
import cv2
import pandas as pd
import os

# CSV 파일 경로
csv_file = 'C:/Users/jk/action_assess_2/alignment/plots/trainer_Z52_user_Z106/base_317_trial_314/max_idx_list.csv'

# 이미지 폴더 경로
base_folder = 'C:/Users/jk/action_assess_2/data/output/frame_key/Z52/317-2-1-15-Z52_D_keypoints'
trial_folder = 'C:/Users/jk/action_assess_2/data/output/frame_key/Z106/314-2-1-15-Z106_D_keypoints'

# CSV 파일 읽기
df = pd.read_csv(csv_file)

# 결과를 저장할 폴더
output_folder = 'C:/Users/jk/action_assess_2/alignment/plots'
os.makedirs(output_folder, exist_ok=True)

shift_amount = 4

# CSV 파일의 각 행에 대해 이미지 결합
for index, row in df.iterrows():

    if index > shift_amount:
        base_image_path = os.path.join(base_folder, f"frame_{int(row['base']-shift_amount-1):04d}.jpg")
        print(f"frame_{int(row['base']):04d}.jpg")
        trial_image_path = os.path.join(trial_folder, f"frame_{int(row['trial']-shift_amount-1):04d}.jpg")
        print(f"frame_{int(row['trial']):04d}.jpg")
        print('-'*100)


        # 이미지 읽기
        base_image = cv2.imread(base_image_path)
        trial_image = cv2.imread(trial_image_path)
        
        # 이미지 결합
        combined_image = cv2.hconcat([base_image, trial_image])
        
        # 결합된 이미지 저장
        output_image_path = os.path.join(output_folder, f"combined_{index}.jpg")
        cv2.imwrite(output_image_path, combined_image)

print("이미지 결합 완료!")
```
