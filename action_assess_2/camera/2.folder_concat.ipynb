```python
import os
from PIL import Image

# 폴더 경로 설정
folder1 = 'C:/Users/jk/action_assess_2/post_data/vector/basic'
folder2 = 'C:/Users/jk/action_assess_2/post_data/vector/world'
output_folder = 'C:/Users/jk/action_assess_2/post_data/vector/concat'

# 출력 폴더가 없으면 생성
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# 폴더1의 이미지 파일 목록 가져오기
images1 = os.listdir(folder1)

for img_name in images1:
    img1_path = os.path.join(folder1, img_name)
    img2_path = os.path.join(folder2, img_name)
    
    # 두 이미지가 모두 존재하는지 확인
    if os.path.exists(img1_path) and os.path.exists(img2_path):
        img1 = Image.open(img1_path)
        img2 = Image.open(img2_path)
        
        # 두 이미지를 수평으로 합치기
        total_width = img1.width + img2.width
        max_height = max(img1.height, img2.height)
        
        new_img = Image.new('RGB', (total_width, max_height))
        new_img.paste(img1, (0, 0))
        new_img.paste(img2, (img1.width, 0))
        
        # 합친 이미지 저장
        new_img.save(os.path.join(output_folder, img_name))
```

```python
import os
from PIL import Image

# 폴더 경로 설정
input_folder = 'C:/Users/jk/action_assess_2/post_data/points'
output_folder = 'C:/Users/jk/action_assess_2/post_data/points'

# 출력 폴더가 없으면 생성
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# 이미지 파일 목록 가져오기
image_files = [f for f in os.listdir(input_folder) if f.startswith('output_image_')]

# 이미지 파일들을 정렬
image_files.sort(key=lambda x: int(x.split('_')[-1].split('.')[0]))

# 이미지들을 가로로 합치기
images = [Image.open(os.path.join(input_folder, img_name)) for img_name in image_files]
total_width = sum(img.width for img in images)
max_height = max(img.height for img in images)

new_img = Image.new('RGB', (total_width, max_height))

current_width = 0
for img in images:
    new_img.paste(img, (current_width, 0))
    current_width += img.width

# 합친 이미지 저장
new_img.save(os.path.join(output_folder, 'concatenated_image.jpg'))
```
